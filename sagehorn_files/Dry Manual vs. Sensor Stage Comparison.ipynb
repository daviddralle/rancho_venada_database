```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["font.sans-serif"] = "Arial"
```

```python

df = pd.read_csv('Stage-Discharge Sagehorn - Sagehorn stage-discharge.csv')
df.index = pd.to_datetime(df['Datetime [YYYY-MM-DD HH:MM]-8'])
#ignore hank, other tribs
dryManual =df[df.Creek=='DRY'] 
#ignore salt discharge


dryManualResampled = dryManual.resample('15Min')
dryManualResampled = dryManualResampled.dropna(subset = ['Staff gauge water level [m]'])
dryManualResampled
```

```python
level4d = pd.read_csv('Data/Output/level4e.csv')
level4d.index=pd.to_datetime(level4d['Datetime UTC - 8'])
level4d.head()
```

```python
merged = level4d.merge(dryManualResampled,left_index=True, right_index=True,how='inner')
merged.head()
```

```python
plt.scatter(merged['Staff gauge water level [m]'],merged['CREEK DRY stage (m)'])
plt.xlabel('Manually-read staff gauge level [m]')
plt.ylabel('Pressure-transducer derived staff gauge level [m]')
plt.title('Dry Creek')
plt.plot([0.4,1.4],[0.4,1.4])
plt.grid()
plt.savefig('Data/Output/OutputFigures/DryCk-Derived-vs-Manual-Stage.pdf')
```

```python
merged['Manual-Derived Difference'] = merged['Staff gauge water level [m]']-merged['CREEK DRY stage (m)']
merged['Manual-Derived Difference'].plot(marker='o')
plt.axhline(0)
plt.ylabel('meters')
plt.legend()
plt.savefig('Data/Output/OutputFigures/DryCk-Derived-Manual-Stage-Difference.pdf')
```
